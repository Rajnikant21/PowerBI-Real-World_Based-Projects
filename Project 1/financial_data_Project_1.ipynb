```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Set random seed for reproducibility
random.seed(42)

# Define parameters
start_date = datetime(2021, 1, 1)
end_date = datetime(2021, 12, 31)
num_transactions = 1000
branch_names = ["Branch A", "Branch B", "Branch C"]
product_categories = ["Electronics", "Clothing", "Books", "Home Appliances"]

# Create a DataFrame to store the data
data = pd.DataFrame(columns=["TransactionDate", "Branch", "ProductCategory", "Revenue", "Expenses"])

# Generate random transactions
for _ in range(num_transactions):
    transaction_date = random.choice(pd.date_range(start_date, end_date))
    branch = random.choice(branch_names)
    category = random.choice(product_categories)
    revenue = round(random.uniform(100, 10000), 2)
    expenses = round(random.uniform(20, 500), 2)

    data = data.append({"TransactionDate": transaction_date,
                        "Branch": branch,
                        "ProductCategory": category,
                        "Revenue": revenue,
                        "Expenses": expenses}, ignore_index=True)

# Save the generated data to a CSV file
data.to_csv("sample_financial_data.csv", index=False)

print("Sample data generation completed and saved to 'sample_financial_data.csv'.")
```
